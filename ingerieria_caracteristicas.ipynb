```python
import pandas as pd

df_partidos = pd.read_csv("datos.csv")
df_partidos.head()
```

```python
# Nueva característica: Diferencia de goles (local - visitante)
df_partidos['diferencia_goles'] = df_partidos['goles_local'] - df_partidos['goles_visitante']

#Ver resultados
df_partidos[['goles_local', 'goles_visitante', 'diferencia_goles']].head()
```

```python
# Nueva característica: ratio de tiros al arco por % de posesión (local)
df_partidos['ratio_tiros_arco_posesion_local'] = df_partidos['tiros_arco_local'] / (df_partidos['posesion_local (%)'])# + 1e-5)

#Ver resultado
df_partidos[['tiros_arco_local', 'posesion_local (%)', 'ratio_tiros_arco_posesion_local']].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#Histogramas de diferencia de goles
sns.histplot(df_partidos['diferencia_goles'], bins=15, kde=True)
plt.title('Distribución de la Diferencia de Goles')
plt.xlabel('Diferencia de Goles (Local - Visitante)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
#Analisis de correlaciones
corr_vars = ['goles_local', 'goles_visitante', 'diferencia_goles', 'ratio_tiros_arco_posesion_local']
corr_matrix = df_partidos[corr_vars].corr()

#Mapa de calor
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')#, fmt=".2f")
plt.title('Matriz de Correlación')
plt.show()
```
