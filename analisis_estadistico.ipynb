```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_partidos = pd.read_csv('datos.csv')
df_partidos.head(10)
```

```python
# Estadistica descriptiva general
df_partidos.describe()
```

```python
# Promedio goles como local y visitante
goles_local = df_partidos[df_partidos['equipo_local'] == 'Cebollitas FC']['goles_local'].mean()
goles_visitante = df_partidos[df_partidos['equipo_visitante'] == 'Cebollitas FC']['goles_visitante'].mean()

print(f'Goles promedio como local: {goles_local}')
print(f'Goles promedio como visitante: {goles_visitante}')
```

```python
# Histogramas goles marcados (local y visitante)
fig, ax = plt.subplots(1, 2, figsize=(12, 4))

sns.histplot(df_partidos['goles_local'], kde=True, ax=ax[0], bins=10)
ax[0].set_title('Distribución de goles como local')

sns.histplot(df_partidos['goles_visitante'], kde=True, ax=ax[1], bins=10, color='orange')
ax[1].set_title('Distribución de goles como visitante')

plt.show()
```

```python
# Boxplotara goles como local
sns.boxplot(x=df_partidos['goles_local'])
plt.title('Boxplot de goles como local')
plt.show() 
```

```python
# Scatterplot de posesión vs goles como local
sns.scatterplot(x='posesion_local (%)', y='goles_local', data=df_partidos)
plt.title('Posesión local vs Goles como local')
plt.show()
```

```python
#Mapa de calor para correlación entre variables clave
plt.figure(figsize=(10, 6))
sns.heatmap(df_partidos[['goles_local', 'goles_visitante', 'posesion_local (%)', 
                         'posesion_visitante (%)', 'tiros_arco_local',
                         'tiros_arco_visitante']].corr(), annot=True, cmap='coolwarm')
plt.title('Mapa de calor - correlación entre variables')
plt.show()
```
